```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[1,2,3,4,5,6]
x2=[1,2,3,4,5,6]
plt.figure(figsize=(9,9))
plt.plot(x, y, label="x" , color="red", linewidth=2, marker='|',markersize=10,markeredgecolor="b")
plt.title('happiness')
x2=np.arange(0,6,1)

plt.plot(x2, x2**2,color="blue",label="W",markersize=2,marker='*')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()

plt.show()
```
